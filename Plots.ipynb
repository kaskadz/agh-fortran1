```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data4 = pd.read_table('kind4.data', sep=r"\s+", header=0)
data8 = pd.read_table('kind8.data', sep=r"\s+", header=0)
data16 = pd.read_table('kind16.data', sep=r"\s+", header=0)
```

```python
data4.sort_values(by='n').plot(figsize=(15, 10), x='n', y='error')
plt.show()
```
